## _Visualización uno_ ##

### How Britain Voted ###

#### Contexto: Año 2017 Irlanda del Norte, el partido conservador perdió su mayoría parlamentaria en las elecciones generales británicas, los votantes produjeron un parlamento sin que ningún partido tenga la mayoría absoluta ####

#### Está visualización nos muestra como se comporto las elecciones y el numero de asientos obtenidos y el cambió frente a las elecciones 2015 donde el partido conservador tenía la mayoría ###

1. $¿qué?$, $¿por qué?$ y $¿cómo?$ 

+ $¿qué?$

Los datos que se quieren mostrar son el comportamiento de los votantes en el Reuino Unido en el año 2017 de las elecciones parlamentarias, en esto sabemos que tenemos la base de datos con datos categoricos ordenados (ubicación geográfica) y tenemos el número de votos obtenidos por el partido ganador en esa ubicación.

+ $¿porqué?$

En particular se quiere visualizar el comportamiento geográfico al menos en esta primera visualización, ya que en esta visualización observamos en que lugares del país tuvo mayoría de votos un partido especifico, está visualización busca presentar como se distribuyeron los votos dentro del territorio del Reino Unido. Adicional a esto se quiere mostrar el cambio entre las elecciones 2015 y 2017 

+ $¿cómo?$

No se puede apreciar en la imagen, pero si nos remitimos a https://www.nytimes.com/interactive/2017/06/08/world/europe/british-general-election-results-analysis.html donde se encuentra completa la visualización podemos observar que cuenta con una interacción sencilla en la cual si nos paramos sobre algún punto nos desgloza el nombre y el porcentaje de votos obtenidos por cada partido en esa zona especifica del total de votos en ese mismo lugar, esto ocurre para las primeras 3 visualizaciones, mientras que para las últimas dos al pararse sobre una barra se despliega información del año, y la representación porcentual que tuvo de votos cada partido en ese año particular.


2. Marcas y canales utilizados

+ Marcas: La marca utilizada en este caso es el area geográfica y las flechas.

+ Canal: El canal para está visualización es el color, el cual nos permite distinguir de un area geográfica a otra que partido fue el que obtivo más votos, además del canal de posición.
